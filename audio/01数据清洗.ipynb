```python
import os


# 文件目录
FilePath = "/Users/qinjianxun/Downloads/audios/archive/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files"
TargetPath = "./data/kaggle"

if not os.path.exists(TargetPath):
    os.makedirs(TargetPath)

# 所有文件名
files = os.listdir(FilePath)

# 所有txt文件
filenames = [f.split('.')[0] for f in files if f.endswith('txt')]

filenames
```

```python
import shutil

if not os.path.exists(os.path.join(TargetPath, '00')):
    os.makedirs(os.path.join(TargetPath, '00'))
if not os.path.exists(os.path.join(TargetPath, '11')):
    os.makedirs(os.path.join(TargetPath, '11'))
if not os.path.exists(os.path.join(TargetPath, '10')):
    os.makedirs(os.path.join(TargetPath, '10'))
if not os.path.exists(os.path.join(TargetPath, '01')):
    os.makedirs(os.path.join(TargetPath, '01'))

# 处理单个文件进行分类
def process_one(filename: str):
    filename_txt = f'{filename}'
    filename_wav = os.path.join(filename, ".wav")
    
    file_path_txt = os.path.join(FilePath, filename_txt)
    file_path_wav = os.path.join(FilePath, filename_wav)

    
    results = []
    with open(file_path_txt) as f:
        lines = f.readlines()
        for l in lines:
             results.append(l.replace('\n', '').replace('\t', ',').split(','))
    
    mid = len(results) / 2
        
    # 判断 第3列 和 第4列 是否过半数
    c3_true_num = 0
    c4_true_num = 0
    
    for result in results:
        if int(result[2]) == 1:
            c3_true_num += 1
        if int(result[3]) == 1:
            c4_true_num += 1
            
   
    if c3_true_num <= mid and c4_true_num <= mid:
        # 00
        shutil.copyfile(file_path_txt, os.path.join(TargetPath, "00", filename_txt))
        shutil.copyfile(file_path_wav, os.path.join(TargetPath, "00", filename_wav))
    elif c3_true_num > mid and c4_true_num > mid:
        # 11
        shutil.copyfile(file_path_txt, os.path.join(TargetPath, "11", filename_txt))
        shutil.copyfile(file_path_wav, os.path.join(TargetPath, "11", filename_wav))
    elif c3_true_num > mid:
        # 01
        shutil.copyfile(file_path_txt, os.path.join(TargetPath, "10", filename_txt))
        shutil.copyfile(file_path_wav, os.path.join(TargetPath, "10", filename_wav))
    elif c4_true_num > mid:
        # 10
        shutil.copyfile(file_path_txt, os.path.join(TargetPath, "01", filename_txt))
        shutil.copyfile(file_path_wav, os.path.join(TargetPath, "01", filename_wav))
        
    print(c3_true_num, c4_true_num, mid)
    
        
        

# process_one('114_1b4_Lr_mc_AKGC417L.txt')
```

```python
# 处理所有文件
for t in filenames:
    process_one(t)
```
